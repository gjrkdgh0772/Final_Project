```python
극안전=pd.read_csv("/content/gdrive/My Drive/FINAL TEAM/파이널프로젝트/비주식팀/Data/비중DF/가격데이터/전체13가격데이터.csv")
```

```python
극안전.Date=pd.to_datetime(극안전.Date)
```

```python
극안전=극안전.set_index('Date')
```

```python
극안전['ksp']=극안전['ksp로그'].cumsum()
```

```python
극안전['ksp']=np.exp(극안전['ksp'])
극안전['ksp']
```

```python
def dd_f(df):
    window=30
    max_in_window=df.rolling(window,min_periods=1).max()
    dd=((df/max_in_window)-1.0)*100
    return dd
```

```python
극안전['kospi']=dd_f(극안전['kospi'])
극안전['유동성추이']=dd_f(극안전['유동성추이'])
극안전['레버리지추이']=dd_f(극안전['레버리지추이'])
극안전['성장성추이']=dd_f(극안전['성장성추이'])
극안전['수익성추이']=dd_f(극안전['수익성추이'])
극안전['시장가치추이']=dd_f(극안전['시장가치추이'])
극안전['활동성추이']=dd_f(극안전['활동성추이'])
극안전['그레이엄추이']=dd_f(극안전['그레이엄추이'])
극안전['린치추이']=dd_f(극안전['린치추이'])
극안전['버핏추이']=dd_f(극안전['버핏추이'])
극안전['삼대투자추이']=dd_f(극안전['삼대투자추이'])
극안전['ksp']=dd_f(극안전['ksp'])
```

```python
극안전['kospi_portfolio']=극안전['kospi']
극안전['kospi']=극안전['ksp']
```

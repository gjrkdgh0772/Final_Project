# 재무비율 분석 정리

## 안정성(유동성) 비율

* 기업의 유동성(단치채무상환능력)과 장기채무상환능력을 나타내는 재무비율
* 유동비율(200%이상적정범위)
* 당좌비율(100%이상적정범위)
* 부채비율(100%이하 안전 200%이상 위험, 절대적기준X)
* 이자보상비율(1보다 크면 문제없지만 3이상이면 좋은기업) 등

## 수익성 비율

* 기업의 경영성과를 나타내는 재무비율
* 매출액순이익률, 총자본이익률
* 자기자본순이익률(ROE),(12%이상우수)(3년간15%이상이면투자할만하다)
* 총자산이익률(ROA)
* 위 비율들은 높을수록 높은기업이라 볼수있다.


## 활동성 비율

* 기업이 보유자산을 얼마나 효율적으로 운영하였는지를 나타내는 재무비율
* 총자산회전율, 매출채권회전율, 재고자산회전율
* 위 비율들은 높을수록 좋은기업이라 볼수있다.
하지만 매출채권회전율,재고자산회전율 같은경우에는 비슷한 상품,유사기업끼리 비교해야한다.

## 성장성 비율

* 기업규모의 증가,이익의 증가 등을 측정하는 비율
* 총자산증가율, 자기자본증가율, 매출액증가율, 순이익증가율
* 위 요소들은 증감율을 기준으로 판단해야 할것이다.

## 시장가치비율

* 기업의 주식가격과 재무제표 항목을 대비시키는 비율
* 주가이익비율 PER
* 주가순자산비율 PBR
* 주가매출액비율 PSR (켄 피셔기준 1보다크면 고평가,1.5이상 피해라, 3이상 매수금지,0.75이하 적극적 탐색)
* 주가현금흐름비율 PCR

## 회사 실적주에 근거한 재무비율
* 자기자본 순 이익률(ROE), 당기순이익률 등에 근거하여 시장대비
overperform 중인 종목 선정.

# 1.시가총액가중방식

* 주식수 * 주가 = 시가총액을 합산한 금액과 기준년도의 시가총액 합산을 비교하여 지수를 산출하는 방식(Valus-weighted index) 

* 샤프에 의해 개발된 이 모델은 개별 회사의 시가총액이 시장에서 차지하는 비중에 따라 투자한다는 것을 가정하고 있다.

* 단점으로는 시가총액이 큰 기업의 주가움직임에 지수가 크게 좌우 된다는 점.

# 2. 동일가중방식

* 선정된 주식 종목을 1억원 기준 n등분 하여 동일하게 가중치를 설정하는 방법 (모든주식에 동일한 금액으로 투자 

* 동일 가중방식은 심리적인 쏠림현상을 억제할수있다.)

# 3.가격가중방식

* 기준시점 주식가격 산술평균과 비교시점 주식가격의 산술편균을 비교하여 수익률을 산출하는 방법, 이방법은 저가주식 가격보다 고가주식 가격이 지수(수익율)에 영향을 많이 준다.

* 투자자들이 모든 주식에 동일한 수량으로 투자하는 경우 가격가중방법이 실효성을 갖는다.

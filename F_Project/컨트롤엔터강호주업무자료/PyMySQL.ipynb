# 테이블생성 로그수익률 5 종목선정

```python
import pymysql

conn = pymysql.connect(host='localhost',
                       user='ctrlenter',
                       password='0000',
                       db='ctrlenter',
                       charset='utf8')

sql = '''CREATE TABLE 로그수익률전체 (
        날짜 date ,
        레버리지기법 float,
        유동성기법 float,
        그레이엄기법 float,
        린치기법 float,
        버핏기법 float,
        삼대투자기법 float,
        성장성기법 float,
        수익성기법 float,
        시장가치기법 float,
        활동성기법 float
        )
        '''

with conn:
    with conn.cursor() as cur:
        cur.execute(sql)
        conn.commit()
```

## 로그수익률전체 넣기

```python
# 2. 데이터베이스 연결
db = pymysql.connect(host='localhost', port=3306, user='ctrlenter', password='0000', db='ctrlenter', charset='utf8')   # charset: 인코딩 설정
# 3. curosr 사용
cursor = db.cursor() 

# 4. 쿼리문 작성
# 컬럼명: 테이블의 컬럼명과 동일하게 입력
# %s 개수: 컬럼 개수만큼 입력
sql = 'INSERT INTO 로그수익률전체(날짜,레버리지기법, 유동성기법,그레이엄기법,린치기법,버핏기법, 삼대투자기법,성장성기법,수익성기법,시장가치기법, 활동성기법) VALUES(%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s )'

# 5. 데이터 import
for idx in range(len(로그수익률전체)):
    cursor.execute(sql, tuple(로그수익률전체.values[idx]))

# 6. 데이터베이스 업데이트 및 작업 종료
db.commit() 

# 7. 데이터베이스 닫기 
db.close() 
```

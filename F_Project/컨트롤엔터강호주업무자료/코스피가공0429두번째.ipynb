```python
def making(per,name):

    per['name_ko'] = name
    per.columns = ['date', 'symbol', 'entity_name', 'type_id', 'report_id', 'MAIN_TERM',
           'value', 'name_ko']
    per = per[['date', 'symbol', 'entity_name', 'type_id', 'report_id', 'MAIN_TERM', 'name_ko','value']]
    
    return per
```

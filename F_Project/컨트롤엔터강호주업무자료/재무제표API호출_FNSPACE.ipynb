# 상장기업목록

```python
import pandas as pd
import time


#최종 결과 데이타를 담을 DataFrame 생성
df_result = pd.DataFrame(columns=['CODE', 'DT'])
print(df_result)

def read_krx_code():
    """KRX로부터 상장기업 목록 파일을 읽어와서 FnSpace용(기업코드 앞에 'A' 추가) DataFrame으로 변경 후 반환"""
    url = 'http://kind.krx.co.kr/corpgeneral/corpList.do?method=download&searchType=13'
    krx = pd.read_html(url, header=0)[0]
    종목 = krx[['종목코드', '회사명']]
    종목 = 종목.rename(columns={'종목코드':'code', '회사명': 'company'})
    종목.code = 종목.code.map('{:06d}'.format)
    for i in range(len(종목)):
        종목['code'][i] = 'A'+종목['code'][i]
    return krx,종목
krx,종목 = read_krx_code()
```
